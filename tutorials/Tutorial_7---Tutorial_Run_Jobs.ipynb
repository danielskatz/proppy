```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import os
os.chdir('..')
import rwpropa as rw
from threading import Thread
```

#### Spherical source with spherical observer

```python
def simulate(file_name):
    sim = rw.PlasmoidSimulation(nr_particles = 10**3, radius = 10**14, energy = 10**15, nr_steps = 1*10**3, diffusion_coefficient = 1*10**21, step_size = 1*10**12)
    %time sim.simulate(file_name)
    sim.plot_arrival_times()
    print('finished: ', file_name)
```

```python
simulate('data/tut_7_test')
```
