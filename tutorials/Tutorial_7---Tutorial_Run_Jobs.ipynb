```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import os
os.chdir('..')
import rwpropa as rw
from threading import Thread
```

#### Spherical source with spherical observer

```python
def simulate(file_name, step_size_diff_factor):
    sim = rw.PlasmoidSimulation(nr_particles = 10**3, radius = 10**14, energy = 10**15, nr_steps = 1*10**3, diffusion_coefficient = 1.5*10**20, step_size = 1*10**12, step_size_diff_factor = step_size_diff_factor)
    %time sim.simulate(file_name)
    sim.plot_arrival_times()
    print('finished: ', file_name)
```

```python
simulate('data/tut_7_test', 1.0)
```

```python
simulate('data/tut_7_test', 10.0)
```
